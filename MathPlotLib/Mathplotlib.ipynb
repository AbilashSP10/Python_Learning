```python
import pandas as pd

data=pd.read_csv("D:\learning\Git_Main\python_learning\Pandas\student_marks.csv")

df=pd.DataFrame(data)

df

df['Total_Marks'] = df['Maths'] + df['Physics'] + df['Chemistry'] + df['English'] + df['Biology'] + df['Economics'] + df['History'] + df['Civics']
df.rename(columns={'Unnamed: 0': 'Student_Name'}, inplace=True)

plt.figure(figsize=(10,5))

plt.bar(df['Student_Name'], df['Total_Marks'],color="brown")

plt.xlabel("Student Name")
plt.ylabel("Total Marks")

plt.title("Student Marks")

plt.xticks(rotation=65)
plt.yticks(rotation=45)

plt.show()


# plt.bar(x)

# plt.show()
```
