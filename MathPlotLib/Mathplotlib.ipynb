```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Sample data

data={
    'Year': [-3, -2, -1, 0, 1, 2, 3],
    'Profit/Loss': [-1000, 2000, -3000, 700, 1500, -1000, 2500]
}

df = pd.DataFrame(data)

#Create a plot

fig, ax = plt.subplots(figsize=(6,6))

#Plot bars

ax.bar(df['Year'], df['Profit/Loss'], color=['green' if val >=0 else 'red' for val in df['Profit/Loss']])

#Set simple limits and fewer ticks

ax.set_xlim(-4,4)
ax.set_ylim(-4000, 4000)

#Set basic grid

ax.set_xticks(range(-4,5))

# ax.set_yticks(range(-4000, 4001, 500))

ticks = plt.yticks()[0]
new_ticks = [tick for tick in ticks if tick != 0]
ax.set_yticks(new_ticks)

ax.grid(True, linestyle='--', linewidth=0.5)

#Move axes to center

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#Axis labels
ax.set_xlabel('Year')
ax.set_ylabel('Profit/Loss')
ax.set_title('Profit and Loss Chart')

fig.savefig("Myplot.png")



plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(-10, 10, 100)
y = x**3

fig, ax = plt.subplots()
ax.plot(x, y)

# Generate Y-tick values in a specific range (e.g., from -1000 to 1000)
# Step size = 200; exclude 0
tick_min = -1000
tick_max = 1000
step = 200

# Using numpy to include both negative and positive ticks
ticks = np.arange(tick_min, tick_max + step, step)
ticks = ticks[ticks != 0]  # remove 0

# Apply custom ticks
ax.set_yticks(ticks)

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate some example 4D data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
w = np.random.rand(100) # The fourth dimension

# Plot the data, using 'w' for color
img = ax.scatter(x, y, z, c=w, cmap='viridis', s=50) # 's' for size
fig.colorbar(img, ax=ax, shrink=0.5, aspect=5) # Add a colorbar

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('4D Scatter Plot (Color represents 4th dimension)')
plt.show()
```
